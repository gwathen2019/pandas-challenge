```python
#import modules
import os
import csv
import pandas as pd
import numpy as np
```

```python
# Make a reference to the purchase_data.csv file path
csv_path = "HeroesofPymoli/Resources/purchase_data.csv"

# Import the purchase_data.csv file as a DataFrame
purchase_df = pd.read_csv(csv_path, encoding="utf-8")
purchase_df.head()
```

```python
purchase_df.shape
purchase_df.info()#Index, Datatype and Memory information
purchase_df.describe()# | Summary statistics for numerical columns
#purchase_df s.value_counts(dropna=False)# | View unique values and counts
#purchase_df.apply(pd.Series.value_counts) # | Unique values and counts for all columns 
```

```python
# Rename the headers to be more explanatory
purchase_df = purchase_df.rename(columns={"SN": "Player Name"})
purchase_df.head()
```

```python
# Calculate the number of unique players in the DataFrame
player_count = len(purchase_df["Player Name"].unique())

print ("Total Players")
player_count
```

```python
player_df = purchase_df.drop_duplicates(subset = "Player Name", keep = 'first')
player_df.describe()
```

```python
#Gender Demographics
    #Percent and Count of Male Players
    #Percent and Count of Female Players
    #Percent and Count of Other/Non-disclosed Players
    
#create gender variables, using the player_df data frame
gender_count = player_df["Gender"].value_counts()
total_count = gender_count.sum()
total_count
player_pct = gender_count/total_count
#create summary table
gender_summary_table = pd.DataFrame({"Total Count": gender_count,
                                    "Pct of Players": player_pct})#["{0:.2f}%.format"]})
gender_summary_table.head()
```

```python
#calculate the number of unique items
items_unique = len(purchase_df["Item ID"].unique())

#calculate the average purchase price
purchase_avg = purchase_df["Price"].mean()

#calculate the total number of purchases
purchase_count = purchase_df["Purchase ID"].count()

#calculate total revenue
total_rev = purchase_df["Price"].sum()

#creat summary table, must pass an index at end
purchase_summary_table = pd.DataFrame({"Number of Unique Items": items_unique,
                                      "Average Price": purchase_avg,
                                      "Total Purchases": purchase_count,
                                      "Total Revenue": total_rev}, index=[0])

purchase_summary_table.head()
```

```python
#Purchase analysis (by gender)
    #Run basic calculations to obtain 
        #purchase count, 
        #avg. purchase price, 
        #avg. purchase total per person etc. by gender
    
    #Create a summary data frame to hold the results
    #Optional: give the displayed data cleaner formatting
    #Display the summary data frame

#group by gender
purchase_gender_df = purchase_df.groupby(["Gender"]).agg(np.sum)
#purchase_gender_df.head()

no_of_purchases_total = purchase_df.groupby(["Gender"]).agg({"Player Name": ['count']})
no_of_purchases_total

avg_purchase_price = purchase_df.groupby(["Gender"]).agg({"Price": ['mean']})
avg_purchase_price

total_purchase_value = purchase_df.groupby(["Gender"]).agg({"Price": ['sum']})
total_purchase_value

avg_purchase_person = purchase_df.groupby(["Gender", "Player Name"]).agg({"Price": ['mean']})
avg_purchase_person = avg_purchase_person.groupby(["Gender"]).mean()
avg_purchase_person

#create summary table
purchase_gender_summary_table = pd.DataFrame({"Purchase Count": no_of_purchases_total,
                                            "Average Purchase Price": avg_purchase_price,
                                            "Total Purchase Value": total_purchase_value,
                                            "Total Avg Purchase per Person": avg_purchase_person})#["{0:.2f}%.format"]})
purchase_summary_table.head(10)
```
